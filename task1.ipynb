```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate fabric-like image
def generate_fabric_image(width, height):
    # Generate smooth color gradients
    x = np.linspace(0, 4 * np.pi, width)
    y = np.linspace(0, 4 * np.pi, height)
    X, Y = np.meshgrid(x, y)
    # Create fabric pattern using sin functions with different frequencies and phases
    R = 0.5 * (1 + np.sin(X))
    G = 0.5 * (1 + np.sin(Y))
    B = 0.5 * (1 + np.sin(X + Y))
    # Scale to 0-255 range
    R = (R * 255).astype(np.uint8)
    G = (G * 255).astype(np.uint8)
    B = (B * 255).astype(np.uint8)
    # Stack color channels to form the image
    fabric_image = np.dstack([R, G, B])
    return fabric_image

# Generate fabric image
image_width = 800
image_height = 600
fabric_img = generate_fabric_image(image_width, image_height)

# Display the fabric image
plt.imshow(fabric_img)
plt.title('Fabric-like Image')
plt.axis('off')
plt.show()

# Calculate mean and standard deviation of the image for each channel
for i, color in enumerate(['Red', 'Green', 'Blue']):
    mean_val = np.mean(fabric_img[:, :, i])
    std_dev = np.std(fabric_img[:, :, i])
    print(f"{color} channel - Mean: {mean_val}, Standard Deviation: {std_dev}")

# Draw histograms for each color channel
colors = ['red', 'green', 'blue']
for i, color in enumerate(colors):
    plt.figure()
    plt.hist(fabric_img[:, :, i].ravel(), bins=256, color=color, alpha=0.7, histtype='stepfilled')
    plt.title(f'Histogram of {color.capitalize()} channel')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.show()
```
